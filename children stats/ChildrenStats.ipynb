```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df1=pd.read_excel("children_stats_only (2).xlsx")
pd.set_option('display.max_columns',None)
df1.head()
```

```python
#to plot bar graph of age 
AgeData=dict(df1.Age.value_counts())
AgeKey = list(AgeData.keys())
AgeValues = list(AgeData.values())
fig = plt.figure(figsize = (5, 4))
# creating the bar plot
plt.bar(AgeKey, AgeValues, color ='blue',width = 0.1)
plt.xlabel("Age")
plt.ylabel("No. of people")
plt.title("Age")
plt.show()
```

```python
#to plot bar graph of gender 
GenderData=dict(df1.Gender.value_counts())
GenderKey = list(GenderData.keys())
GenderValues = list(GenderData.values())
fig = plt.figure(figsize = (2, 2))
# creating the bar plot
plt.bar(GenderKey, GenderValues, color ='blue',width = 0.1)
plt.xlabel("Gender")
plt.ylabel("No. of people")
plt.title("Gender")
plt.show()
gender=df1.Gender.value_counts()
gender
```

```python
#to plot bar graph of marital status 
MaritalStatus=dict(df1.MaritalStatus.value_counts())
MaritalStatusKey = list(MaritalStatus.keys())
MaritalStatusValues = list(MaritalStatus.values())
fig = plt.figure(figsize = (3, 3))
# creating the bar plot
plt.bar(MaritalStatusKey, MaritalStatusValues, color ='blue',width = 0.1)
plt.ylabel("No. of people")
plt.title("Marital Status")
plt.show()
Marital_Status=df1.MaritalStatus.value_counts()
Marital_Status
```

```python
DietStatus=dict(df1.Diet.value_counts())
DietStatusKey = list(DietStatus.keys())
DietStatusValues = list(DietStatus.values())
fig = plt.figure(figsize = (8, 3))
# creating the bar plot
plt.bar(DietStatusKey, DietStatusValues, color ='blue',width = 0.1)
plt.xlabel("Diet")
plt.ylabel("No. of people")
plt.title("Diet Status")
plt.show()
Diet_Status=df1.Diet.value_counts()
Diet_Status
```

```python
xWeight = list(df1['Weight'])
yHeight = list(df1['Height'])
plt.figure(figsize=(5,5))
plt.style.use('seaborn')
plt.scatter(xWeight,yHeight)
plt.title("Height and weight of students Scatter Plot")
plt.xlabel("Weight")
plt.ylabel("Height")
plt.show()
```

```python
Occupation_data=df1['Occupation'].value_counts().tolist() 
Occupation_label=df1['Occupation'].value_counts().index 
plt.pie(Occupation_data,labels=Occupation_label,autopct='%1.2f%%',textprops={'size':10}, pctdistance=0.9)  
plt.title("Occupation of students")
plt.show()
```
