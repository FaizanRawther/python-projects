```python
from bokeh.plotting import figure,show,output_file
import pandas as pd
```

```python
df1=pd.read_excel("https://github.com/pythonizing/data/raw/master/verlegenhuken.xlsx")
df1
```

```python
df1["Temperature"]=df1["Temperature"]/10
df1["Pressure"]=df1["Pressure"]/10
```

```python
output_file=("plotting weather")
f1=figure(plot_width=500,plot_height=400,tools='pan')
f1.title.text="Temperature and air Pressure"
f1.title.text_color="Gray"
f1.title.text_font="times"
f1.title.text_font_style="bold"
f1.xaxis.axis_label="Temperature(C)"
f1.yaxis.axis_label="Pressure(hPa)"
f1.circle(df1["Temperature"],df1["Pressure"], size =0.5)
show(f1)
```
